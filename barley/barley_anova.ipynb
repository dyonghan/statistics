```python
def avg_std(df):
    avg = df.mean(axis=0)
    std = df.std(axis=0)
    avg_std = pd.concat([avg, std], axis=1, keys=['avg', 'std'])
    display(avg_std)
```

```python
def oneway(df):
    size = len(df.keys())
    mx = np.zeros([size,size], np.float32)
    mx = pd.DataFrame(mx, index=df.keys(), columns=df.keys())
    for k1 in df.keys():
        for k2 in df.keys():
            Fv, pv = stats.f_oneway(df[k1], df[k2])
            mx[k1][k2] = pv
    display(mx)
    display('p <= 0.05:')
    display(mx<=0.05)
    display('p <= 0.01:')
    display(mx<=0.01)
```
