```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("feature_code.csv", index_col=0)
```

```python
for r in df.index:
    nums = list()
    end = False
    for i, c in enumerate(df.columns):
        if pd.isnull(df.loc[r,c]):
            end = True
        elif end:
            print('Error: data after end, [{},{}]'.format(r, c))
        elif i%2!=0:
            n = int(df.loc[r,c])
            if not nums:
                nums.append(n)
            elif n == nums[-1] + 1:
                nums.append(n)
            else:
                print('Error: num wrong, [{},{}]'.format(r, c))
print('checked feature code')
```

```python
feature_code = dict()
for r in df.index:
    feature_code[r] = dict()
    for i, c in enumerate(df.columns):
        if pd.isnull(df.loc[r,c]):
            break
        elif i%2 == 0:
            k = df.loc[r,c]
        else:
            feature_code[r][k] = int(df.loc[r,c])
print(feature_code)
```

```python
def to_num(df):
    for r in df.index:
        if not feature_code[r]:
            df.loc[r] = pd.to_numeric(df.loc[r])
        else:
            print(r, feature_code[r])
            df.loc[r] = df.loc[r].apply(lambda x: x if pd.isnull(x) else feature_code[r][x.strip()])
    return df
```

```python
jiukeng = pd.read_csv("tea-jiukeng.csv", index_col=0)
jiukeng = to_num(jiukeng)
```

```python
jiukeng.loc['完全开放花冠大小1'] = (jiukeng.loc['完全开放花冠大小1'] + jiukeng.loc['完全开放花冠大小2'])/2
jiukeng = jiukeng.rename({'完全开放花冠大小1':'完全开放花冠大小'})
jiukeng = jiukeng.drop(index=['完全开放花冠大小2'])
```

```python
keys = ['萼片数','完全开放花冠大小','花瓣数','最大花瓣质地','花柱长度','柱头裂位','雌雄蕊相对高度','着生状态','叶长cm','叶宽cm','叶正面颜色','叶正面隆起性','叶形','叶身形态','叶基形态','叶缘形态']
removed_keys = set(jiukeng.index) - set(keys)
jiukeng = jiukeng.drop(removed_keys)
```

## Correlation

```python
def crowd(df):
    short_rows = list()
    for r in df.index:
        l = list()
        for c in df.columns:
            if not pd.isnull(df.loc[r,c]):
                l.append(df.loc[r,c])
        if len(l) < 30:
            short_rows.append(r)
        l.extend([np.nan for _ in range(df.shape[1]-len(l))])
        df.loc[r] = l
    return df, short_rows
```

```python
df, short_rows = crowd(jiukeng)
df = df.drop(short_rows)
print('drop', short_rows)
df.to_csv('results/jiukeng_num.csv')
df = df.dropna(axis=1)
df = df.T
```

```python
drop_cols = list()
for c in df.columns:
    if df[c].std() == 0:
        drop_cols.append(c)
print('drop', drop_cols)
df = df.drop(drop_cols, axis=1)
df
```

```python
df_corr = pd.DataFrame()
for r, row in enumerate(df.columns[1:]):
    for c, col in enumerate(df.columns):
        if r >= c:
            rv, pv = stats.pearsonr(df[col], df[row])
            df_corr.loc[row,col] = ('%.3f*' if pv<=0.05 else '%.3f') % rv
# pd.set_option('precision', 3)
display(df_corr)
```

```python
df_corr.to_csv('results/jiukeng_corr.csv')
```

## 方差分析

```python
def avg_std(df):
    avg = df.mean(axis=0)
    std = df.std(axis=0)
    avg_std = pd.concat([avg, std], axis=1, keys=['avg', 'std'])
    avg_std.to_csv('results/jiukeng_avg_std.csv')
    display(avg_std)
```

```python
def oneway(df):
    size = len(df.keys())
    mx = np.zeros([size,size], np.float32)
    mx = pd.DataFrame(mx, index=df.keys(), columns=df.keys())
    for k1 in df.keys():
        for k2 in df.keys():
            Fv, pv = stats.f_oneway(df[k1], df[k2])
            mx[k1][k2] = pv
    display(mx)
    display('p <= 0.05:')
    df05 = mx<=0.05
    df05.to_csv('results/jiukeng_0.05.csv')
    display('p <= 0.01:')
    df01 = mx<=0.01
    df01.to_csv('results/jiukeng_0.01.csv')
```
