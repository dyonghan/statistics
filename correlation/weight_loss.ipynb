```python
import pandas as pd
df = pd.read_csv("sorghum/weight_loss.csv")
```

```python
df = df.dropna()
df = df.loc[:, 'Var 1_72h':'Var 6_144h']
df = df.apply(lambda x: x.astype(str).str.replace(',','.').astype(float))
display(df)
```

```python
import numpy as np
from scipy import stats

def one_way(df1):
    mx = pd.DataFrame(np.zeros([6,6], np.float32), index=df1.keys(), columns=df1.keys())
    for k1 in df1.keys():
        for k2 in df1.keys():
            Fv, pv = stats.f_oneway(df1[k1], df1[k2])
            mx[k1][k2] = pv
    display(mx)
    display('p <= 0.05:')
    display(mx<=0.05)
    display('p <= 0.01:')
    display(mx<=0.01)
```

```python
df1 = df.loc[:, 'Var 1_72h':'Var 6_72h']
one_way(df1)
```

```python
df2 = df.loc[:, 'Var 1_144h':'Var 6_144h']
one_way(df2)
```
