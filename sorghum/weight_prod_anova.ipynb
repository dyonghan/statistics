```python
import pandas as pd
df = pd.read_csv("sorghum/weight_prod.csv")
# every 4 elements belong to a variety
display(df)
```

```python
import numpy as np
from scipy import stats

def one_way(column):
    df1 = pd.DataFrame(np.zeros([4,6], np.float32), columns=[str(x+1) for x in range(6)])
    i = 0
    for c in range(6):
        for r in range(4):
            df1.iloc[r, c] = df.iloc[c*4 + r,column]
    display(df1)

    mx = pd.DataFrame(np.zeros([6,6], np.float32), index=df1.keys(), columns=df1.keys())
    for k1 in df1.keys():
        for k2 in df1.keys():
            Fv, pv = stats.f_oneway(df1[k1], df1[k2])
            mx[k1][k2] = pv
    display(mx)
    display('p <= 0.05:')
    display(mx<=0.05)
    display('p <= 0.01:')
    display(mx<=0.01)
```
